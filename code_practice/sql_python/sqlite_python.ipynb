#sql + python

```python
#imports
import sqlite3
```

```python
#connection
try:
    database = sqlite3.connect(database="db_1.db")
except Exception as e:
    print(e)
else:
    print(database)
```

```python
#database + CRUD

try:
    cursor = database.cursor()
    
    #table
    cursor.execute("""CREATE TABLE IF NOT EXISTS stock_market 
                   (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, tag text unique)"""
                   )
    
    #clear to run
    cursor.execute("""DELETE FROM STOCK_MARKET""")
    database.commit()
    
    #data
    cursor.execute("""INSERT INTO stock_market (name,tag) VALUES ("AAPL34", "APPLE STOCK")""")
    cursor.execute("""INSERT INTO stock_market (name,tag) VALUES ("AMZO34", "AMAZON STOCK")""")
    
    #read - select
    result = cursor.execute("""SELECT ID, NAME, TAG FROM STOCK_MARKET""").fetchall()
    print(result)
    for r in result:
        print(f"Id: {r[0]} -> Name: {r[1]} -> Tag: {r[2]}\n")
    
    #update
    cursor.execute("""UPDATE STOCK_MARKET SET name = 'STOCK_UPDATE_TEST' WHERE name = 'AMZO34'""")
    database.commit()
    select_update = cursor.execute("""SELECT * FROM STOCK_MARKET WHERE name = 'STOCK_UPDATE_TEST'""").fetchall()
    for r in select_update:
        print(f"update: {r}")
        
    #delete
    cursor.execute("""DELETE FROM STOCK_MARKET WHERE name = 'STOCK_UPDATE_TEST'""")
    delete_result = cursor.execute("""SELECT * FROM STOCK_MARKET""").fetchall()
    for r in delete_result:
        print(f"delete result: {r}")
    
except Exception as e:
    print(e)
else:
    database.commit()
    database.close()
```
