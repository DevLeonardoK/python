## Section 5 — Adding and Removing Columns

**Exercise 5.1** — Add a new column `salary_with_bonus` equal to `salary * 1.10` (10% bonus).


````python
# your code here
df["salary_with_bonus"] = df["salary"] * 1.10
df.head()
````

**Exercise 5.2** — Add a column `seniority` with value `'Senior'` if `years_at_company >= 10`, else `'Junior'`.

````python
import numpy as np
# your code here

rows = df.shape[0]
df["seniority"] = ""

for r in range(rows):
    if df.iloc[r].years_at_company >= 10:
        df.loc[r,"seniority"] = "Senior"
    else:
        df.loc[r,"seniority"] = "Junior"
    
df.head()
````

**Exercise 5.3** — Remove the `salary_with_bonus` column you just created.

````python
# your code here
df.drop(columns="salary_with_bonus")
````

**Challenge 10.3** — Create a summary table showing, for each `city`:
- Total number of employees
- Average salary
- Average performance rating

````python
# your code here
df_basic = pd.DataFrame()

total_number = df.groupby("city")["id"].count()

average = df.groupby("city")["salary"].mean()

performance_rating_av = df.groupby("city")["performance_rating"].mean()

df_basic["total"] = total_number

df_basic["salary average"] = average

df_basic["performance average"] = performance_rating_av

df_basic
````
