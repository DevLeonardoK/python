```python
#code clean
planilha_null = pd.read_csv("funcionarios_nulos.csv")
planilha_null.info()
```

```python
#remove only series NaN
planilha_null.dropna(subset="salario")
```

```python
#remove if some columns values is Nan
planilha_null.dropna()
```

```python
#complete NaN values with mean, others

salario_mean = planilha_null["salario"].mean()

planilha_null["salario"] = planilha_null["salario"].fillna(salario_mean)

planilha_null
```

```python
#change data type

planilha_null["salario"] = planilha_null["salario"].astype(int)
planilha_null.head()
```
