```python
#Read excel
planilha = pd.read_csv('funcionarios.csv')

#Show first 10 lines
planilha.head(10)

#Show last 10 lines
planilha.tail(10)
```

```python
#Read some columns
planilha[["nome","departamento"]]
```

```python
#series validation
planilha["salario"] > 8000
```

```python
#data filter
planilha[planilha["salario"]> 10000]
```

```python
#mutiple filters

salario_filter = planilha["salario"] > 12000
idade_filter = planilha["idade"] > 25

planilha[salario_filter & idade_filter]
```

```python
#add column and inserting

planilha["salario bonus"] = planilha["salario"] * 1.5
planilha
```

```python
#remove the column

planilha.drop("salario bonus", axis=1) #name and column
```
